```python
import time

current_time_ns = time.time_ns()
print(current_time_ns//1000)

inicio = time.perf_counter()
print(f"inicioa  = {inicio}")
# Coloque aqui o código que você deseja medir
time.sleep(1.2) # Aguarda 1.2 segundos como exemplo
fim = time.perf_counter()
print(f"fim = {fim}")

tempo_em_ms = (fim - inicio) * 1000
print(f"Tempo decorrido: {tempo_em_ms:.2f} ms")

print(time.perf_counter())
print(time.time())
print(time.time_ns())
```
